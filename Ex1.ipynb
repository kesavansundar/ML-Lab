```python
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
data={"Name":['Rex','Peter','Sam','Rita','Hilda'],"Age":[20,12,50,42,35],"Height":[120,150,142,153,145],"Weight":[50,42,57,45,75],"BP":[80,100,70,85,91]}
frame=DataFrame(data)
print(frame)
frame.to_csv("Data.csv")
df=pd.read_csv("Data.csv")
df=df.replace(0,df.mean(numeric_only=True))
print(df)
frame2=df[["Age","Height","Weight","BP"]]
print(frame2)
scaler = StandardScaler()
scaler.fit(frame2)
scaled=scaler.transform(frame2)
print(scaled)
sel=VarianceThreshold(threshold=(.5*(1-0.9)))
useful=sel.fit_transform(scaled)
print(useful)
frame3=df[["Age","Height","Weight","BP"]]
min_max=MinMaxScaler()
min_max.fit(frame3)
scaled2=min_max.transform(frame3)
print(scaled2)
useful1=sel.fit_transform(scaled2)
print(useful1)
```
